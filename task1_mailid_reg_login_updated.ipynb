```python
# email registration
# condition 1 mail id should not start with spl character and numbers
# condition 2 @ and '.' should be present but not consecutively
# condition 3 username followed by @
# condition for creating one email id and store that id in a file
# this task has to be accomplished using python alone. No web.


import re       # regular expression
import getpass  # for security purpose we need to cover the password


# whether the username present in txt file
# if yes then retrieve the password
# if no print wrong username(username not present)
def forgot_password_checkforid(x):      # stage 9
    print('Enter Username, word before "@" is the username (eg.<Username>@new.com)')
    username = input()
    f1 = open('mail_ids1.txt', 'r')
    count = 0
    for l_no, line in enumerate(f1):
        if username in line:
            print('exist')
            line_no = l_no
            user_db = line
            user_db = user_db[0:(len(user_db))]
        count += 1
    f1.close()

    if line_no != (count-1):
        user_db = user_db[0:(len(user_db)-1)]

    user_db = user_db[0:(len(user_db))]

    whole_line = user_db
    whole_line = whole_line.split(',')
    pswd_id = whole_line[len(whole_line)-1]
    user_db = user_db.split('@')
    user_db = user_db[0]

    if user_db == username:
        print("Correct Username your password has been retrieved")
        print("Your Original Password is ", pswd_id)

    elif user_db != username:
        print("Username not present")


# Only if the password is wrong
# whether try again or retrieve password
def forgot_password(a):      # stage 8
    if a == 'F' or a == 'f':
        forgot_password_checkforid(True)    # create new password
    elif a == 'T' or a == 't':
        log_in(True)    # try again
    else:
        print("Please enter either 'F' or 'T' ")
        in_press = input()
        forgot_password(in_press)

# only if log in chosen rather than register
# check the presence of mail id and password in txt file
# if yes continue further
# if no go to forgot password stage
def log_in(x):  # stage 7
    print('Enter mail id:', end="")
    id1 = input()
    # print('Enter Password: ', end="")
    password = getpass.getpass()
    check = id1 + ',' + password

    f1 = open('mail_ids1.txt', 'r')
    # print('id1', id1)
    # print('passw', password)
    # print('chek', check)
    # print(type(check))
    # print(len(check))
    id_db = 0
    count = 0
    for l_no, line in enumerate(f1):
        if id1 in line:
            print('exist')
            line_no = l_no
            id_db = line

        count += 1
    f1.close()
    if line_no != (count-1):
        id_db = id_db[0:(len(id_db)-1)]
    id_db = id_db[0:(len(id_db))]

    if check == id_db:
        print("Logged in successfully")

    else:
        print("Incorrect password or mail id")
        print('Forgot Password?')
        print("Press 'T' for Try again or Press 'F' create new password: ")
        type1 = input()
        forgot_password(type1)


# store the data in a text file
def file1_handle(id, pswd):     # stage 6
    file1 = open("mail_ids1.txt", "a")
    file1.write("\n")
    file1.write(id)
    file1.write(",")
    file1.write(pswd)
    file1.close()
    print("Email id successfully stored in our Data Base")
    return file1


# only for registration purpose
# get the id and password and store them in a txt file
def password_operation(pswd, email):   # stage 5
    password = pswd
    size_password = len(password)
    upper, lower, num, spl, space = 0, 0, 0, 0, 0
    spl_e = []
    if 5 < size_password < 16:
        for i in range(len(password)):
            if password[i] >= 'A' and password[i] <= 'Z':
                upper += 1
            elif password[i] >= 'a' and password[i] <= 'z':
                lower += 1
            elif '0' < password[i] < '9':
                num += 1
            elif password[i] == ' ':
                space += 1
            else:
                spl += 1
                spl_e.append(password[i])
        if upper > 0 and lower > 0 and num > 0 and spl > 0 and space == 0:
            '''print('upper: ', upper)
            print('lower: ', lower)
            print('num: ', num)
            print('spl: ', spl)
            print('space: ', space)
            print(spl_e)'''
            print("Email id created successfully")
            file1_handle(email, password)
        else:
            print("invalid password, pls enter a valid password")
            print("And use atleast 1 uppercase, 1 lowercase, 1 number, 1 special character without any spaces")
    else:
        print("password length should be in the range of 5 to 16")
        print("And use atleast 1 uppercase, 1 lowercase, 1 number, 1 special character without any spaces")


# only for registration
# check all the credentials are followed
# credentials mentioned as comment in the top
def check(email):   # stage 4
    email_cond = '[A-Za-z_]+[A-Za-z0-9_+.-_]+@[A-Za-z0-9]+.[A-Z|a-z]{2,}'
    if re.match(email_cond, email):
        # print("valid email")

        # password enter
        print("Now enter password length should be in the range of 5 to 16")
        print("And use atleast 1 uppercase, 1 lowercase, 1 numeric value, 1 special character without any spaces")
        password1 = getpass.getpass()
        return password_operation(password1, email)
    else:
        print("invalid email")


# if registration chosen in the previous stage
# get mail id as input
# go for a check, to meet the restrictions
def registration(x):    # stage 3
    email = input("Enter Email: ")
    check(email)


# in order to choose whether wants to login or register
# if yes, register then go to registration
# if yes, login then go to log in page
def login_or_register(x):  # stage 2
    if x == 'R' or x == 'r':
        print('Register')
        registration(True)      # join register

    elif x == 'L' or x == 'l':
        print('Login')
        log_in(True)  # join log in

    else:
        print("Not applicable, please enter either 'R' or 'L' ")
        again = input()
        login_or_register(again)


# stage 1
# welcoming
# take the input and start the desired process
print("welcome to project of email")
print("Press 'R' for Register, or Press 'L' for log in: ", end='')
press = input()
login_or_register(press)
```
